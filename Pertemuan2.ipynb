```python
df.describe(include=['object'])
```


<h3>Value Counts</h3>
<p>Penghitungan nilai adalah cara yang baik untuk memahami berapa banyak unit dari setiap karakteristik/variabel yang kita miliki. Kita bisa menerapkan metode "value_counts" pada kolom 'drive-wheels'. Jangan lupa metode "value_counts" hanya berfungsi pada seri Pandas, bukan Pandas Dataframes. Sehingga, kita hanya menyertakan satu tanda kurung "df['drive-wheels']" bukan dua tanda kurung "df[['drive-wheels']]".</p> 


```python
df['jumlah'].value_counts()
```

Kita dapat melakukan konversi kedalam dataframe dengan cara sebagai berikut :


```python
df['jumlah'].value_counts().to_frame()
```

Mari ulangi langkah di atas tetapi simpan hasilnya ke dataframe "drive_wheels_counts" dan ganti nama kolom 'drive-wheels' menjadi 'value_counts'. 


```python
drive_wheels_counts = df['jumlah'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'jumlah': 'value_counts'}, inplace=True)
drive_wheels_counts
```

Sekarang mari kita ganti nama indeks menjadi 'drive-wheels': 


```python
drive_wheels_counts.index.name = 'jumlah'
drive_wheels_counts
```

Kita dapat mengulangi proses di atas untuk variabel 'engine-location'. 


```python
# engine-location as variable
engine_loc_counts = df['jumlah'].value_counts().to_frame()
engine_loc_counts.rename(columns={'jumlah': 'value_counts'}, inplace=True)
engine_loc_counts.index.name = 'jumlah'
engine_loc_counts.head(10)
```

```python
#Tugas Praktikum Jobsheet 1
from numpy import asarray
import pandas as pd
import string

data = pd.read_csv
with open('/content/drive/MyDrive/Colab Notebooks/DataSet/orange.txt', 'r') as data:
    text = data.read()
    words = text.split()
    table = str.maketrans("","", string.punctuation)
    stripped = [w.translate(table) for w in words]
    assembled = " ".join(stripped)
    corpus = [
        assembled
    ]

    from sklearn.feature_extraction.text import TfidfVectorizer

    vectorizer = TfidfVectorizer(stop_words='english')
    response = vectorizer.fit_transform(corpus)
    print(response)
```
